```python
from simpletransformers.classification import MultiLabelClassificationModel
import pandas as pd
import sys, time
```

```python
data_path = "指導的學生/2019_吳玟萱/2019_1104_Joke_Datasets/mlabel_corpora/JokeSkill.txt"
train_random_path = "指導的學生/2019_吳玟萱/2019_1104_Joke_Datasets/mlabel_corpora/JokeSkill_train_random.txt"
```

```python
# set global variables: df
df = pd.read_csv(data_path, delimiter="\t")
#data_raw = df.loc[np.random.choice(data_raw.index, size=2000)]
train_random = pd.read_csv(train_random_path, delimiter="\t")
print(df.shape, train_random.shape) # same as data_raw.shape in Jupyter
```

```python
from sklearn.model_selection import train_test_split
# ID=L1850 為分界，之前：吳玟萱，之後：黃亭筠，均為中文系同一屆
train, test = train_test_split(df, train_size=1691, shuffle=False) 
# (tempararily) set global variables: train, test 
print(train_random.shape)
print(train.shape, test.shape)
print(train.columns)
print(train_random.columns)
```

## set global variables: train_df, test_df

```python
# https://stackoverflow.com/questions/34682828/extracting-specific-selected-columns-to-new-dataframe-as-a-copy
# selected column index: 1, 6, 7
# new = old.iloc[: , [1, 6, 7]].copy() 
# new = old[['A', 'C', 'D']].copy()

train_df = train_random[['Content']].copy()
#train_df = train_df.drop(labels = ['ID', 'Title', 'Content'], axis=1)
#https://stackoverflow.com/questions/33098383/merge-multiple-column-values-into-one-column-in-python-pandas
train_df['labels'] = train_random[train_random.columns[3:]].apply(
    lambda x: list(x), axis=1)
#    lambda x: ','.join(x.dropna().astype(str)), axis=1)
#train_df['labels'] = train_df[['labels']].apply(lambda x: x.split(','), axis=1)
# train.columns = ['ID', 'Title', 'Content', 'Pun', 'Exaggeration', 'Anthropomorphism', 'Bridge_Inference', 'Illogical', 'Irony', 'Imitation', 'Others']
#train_df = train_df.drop(labels = train.columns, axis=1)


test_df = test.loc[:, ['Content']].copy()
#https://stackoverflow.com/questions/33098383/merge-multiple-column-values-into-one-column-in-python-pandas
test_df['labels'] = test.loc[:, test.columns[3:]].apply(
    lambda x: list(x), axis=1)
#    lambda x: ','.join(x.dropna().astype(str)), axis=1)
#test_df['labels'] = test_df[['labels']].apply(lambda x: x.split(','), axis=1)
#test_df = test_df.drop(labels = test.columns, axis=1)

print(f'{train_random.shape}, {test.shape}, {train_df.shape}, {test_df.shape}')
print()
print(train_df.head())
print(test_df.head())
```

```python
# Train and Evaluation data needs to be in a Pandas Dataframe containing at least two columns, a 'text' and a 'labels' column. The `labels` column should contain multi-hot encoded lists.

#train_data = [['Example sentence 1 for multilabel classification.', [1, 1, 1, 1, 0, 1]]] + [['This is another example sentence. ', [0, 1, 1, 0, 0, 0]]]
#train_df = pd.DataFrame(train_data, columns=['text', 'labels'])
#train_df = pd.DataFrame(train_data)

#eval_data = [['Example eval sentence for multilabel classification.', [1, 1, 1, 1, 0, 1]], ['Example eval senntence belonging to class 2', [0, 1, 1, 0, 0, 0]]]
#eval_df = pd.DataFrame(eval_data)

#print(train_df.head())
#print(eval_df.head())
```

```python
# Create a MultiLabelClassificationModel
model = MultiLabelClassificationModel('bert', 'bert-base-chinese', 
#model = MultiLabelClassificationModel('bert', '/Users/sam/GoogleDrive/BERT/bert/chinese_L-12_H-768_A-12/', 
    num_labels=8, use_cuda=True, 
    args={'reprocess_input_data':True, 'overwrite_output_dir':True, 'num_train_epochs':10,
          'max_seq_length': 512, 'train_batch_size': 8, 'eval_batch_size': 8,
          'output_dir':'outputs/Skill_random'})
# You can set class weights by using the optional weight argument
```

```python
# https://ipywidgets.readthedocs.io/en/stable/user_install.html
# conda update -n base -c defaults conda # this will update conda version for all virtualenv
# conda install -c conda-forge ipywidgets

stime = time.time()
# Train the model
model.train_model(train_df)
print('training time: %1.2f'%(time.time()-stime))
```

```python
stime = time.time()
# Evaluate the model
result, model_outputs, wrong_predictions = model.eval_model(test_df)
print(result)
print(model_outputs) # this is the raw prediction
print('Wrong Prediction:')
print(wrong_predictions)
print('Eval time: %1.2f'%(time.time()-stime))
```
