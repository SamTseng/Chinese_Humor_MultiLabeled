```python
import codecs

oLines = None
#with codecs.open( "JokeSkill.txt", "r", "utf-8" ) as f:
#with codecs.open( "JokeSkill_train.txt", "r", "utf-8" ) as f:
with codecs.open( "JokeSkill_test.txt", "r", "utf-8" ) as f:
    oLines = f.readlines()

jokeContent = {}
for line in oLines:
    l = line.strip()
    arr = l.split( "\t" )
    sLen = len( arr[2] )
    if  sLen  not in jokeContent:
#    if str( sLen ) not in jokeContent:
#        jokeContent[ str( sLen ) ] = []
        jokeContent[  sLen  ] = []
#    jokeContent[ str( sLen ) ].append( l )
    jokeContent[ sLen  ].append( l )

print("Number of different Content length: %4d"%len(jokeContent.keys()))
#print(sorted(jokeContent.keys()))

sum = 0
for i in range( 0, 2026 ):
    if i % 10 != 0:
#        k = str(i)
        k = i
        val = len( jokeContent[k] ) if k in jokeContent else 0
        sum += val
    else:
        if (i > 500 and i < 545) or ( i > 545 and i < 655 ) or ( i > 655 and i < 750 ) or ( i > 750 and i < 2025 ):
            continue

        print( "%d\t%d" % ( i, sum ) )
        sum = 0
```
